```
x = []
for i in range(len(df['Content'])) :
  x.append(df['Content'][i])
print(x)
```

```
x2 =  ','. join(str(v) for v in x)
```

```
#출력을 위해서 임시 class 
class List(list): 
    def __str__(self): 
        return "[" + ", ".join(["%s" % x for x in self]) + "]"
```

```
from konlpy.tag import Okt


okt = Okt()
okt_tokens = okt.morphs(x2)


print(List(okt_tokens))
```

```

oktTag = []
for i in okt_tokens:
  oktTag += okt.pos(i)
print(oktTag[0:])
```

```
StopPos = ['Suffix', 'Punctuation', 'Josa', 'Foreign', 'Alpha', 'Number']
StopWord = ['의', '이', '를', '은', '과', '수', '했다', '것', '있는', '한다', '하는', '그', '있다', '는', '가', '인', '.', '을']
```

```
word = []
for tag in oktTag:
  if tag[1] not in StopPos:
    if tag[0] not in StopWord:
      word.append(tag[0])
```
